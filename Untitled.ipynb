```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('List of Countries by Sugarcane Production.csv')
```

```python
data[data["Yield (Kg / Hectare)"].isnull()]
```

Removing the null values which we found

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

#### Handling the data errors incountered

```python
data['Production (Tons)'] = data['Production (Tons)'].str.replace('.','')
data['Acreage (Hectare)'] = data['Acreage (Hectare)'].str.replace('.','')

data['Production per Person (Kg)'] = data['Production per Person (Kg)'].str.replace('.','').str.replace(',','.')
data['Yield (Kg / Hectare)'] = data['Yield (Kg / Hectare)'].str.replace('.','').str.replace(',','.')
```

```python
data.head()
```

#### Renaming the attributes

```python
data.rename(columns = {"Production (Tons)":"Production(Tons)",
                       "Production per Person (Kg)":"Production/Person(Kg)",
                       "Acreage (Hectare)":"Acreage(Hectare)",
                       "Yield (Kg / Hectare)":"Yield(Kg/Hectare)"}, inplace = True)
```

```python
data.head()
```

#### Removing Unnamed: 0 columns

```python
data.drop("Unnamed: 0",axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.dtypes
```

#### Changing the data type for the selected columns from object to float

```python
convert_col = ["Production(Tons)","Production/Person(Kg)","Acreage(Hectare)","Yield(Kg/Hectare)"]
data[convert_col] = data[convert_col].astype(float)
```

```python
top_12_countries = data[["Country", "Production/Person(Kg)", "Acreage(Hectare)"]].set_index("Country").sort_values("Production/Person(Kg)", ascending=False).head(12)

# Create the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(top_12_countries["Acreage(Hectare)"], top_12_countries["Production/Person(Kg)"])

# Add labels and title
plt.xlabel("Acreage (Hectare)")
plt.ylabel("Production/Person (Kg)")
plt.title("Top 12 Countries by Production/Person (Kg)")

# Annotate points with country names
for country, x, y in zip(top_12_countries.index, top_12_countries["Acreage(Hectare)"], top_12_countries["Production/Person(Kg)"]):
    plt.text(x, y, country, fontsize=9)
```
